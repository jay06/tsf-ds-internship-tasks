# KMeans Clustering algorithm using IRIS dataset

```python
#import the packages
from sklearn import datasets
from sklearn.cluster import KMeans
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Load the iris data and take a quick look at the structure of the data
iris = datasets.load_iris()
print(iris.data)
print(iris.target)
```

```python
#Letâ€™s convert our arrays to a pandas DataFrame for ease of use
x = pd.DataFrame(iris.data, columns=['Sepal Length', 'Sepal Width', 'Petal Length', 'Petal Width'])
y = pd.DataFrame(iris.target, columns=['Target'])
```

## Visualise the data in a scatter plot to see if there is any pattern visible

```python
colors = np.array(['red', 'green', 'blue'])
plt.scatter(x['Sepal Length'], x['Sepal Width'], c=colors[y.Target], s=40)
plt.title('Sepal Length vs Sepal Width')
plt.show()
```

```python
plt.scatter(x['Petal Length'], x['Petal Width'], c= colors[y.Target], s=40)
plt.title('Petal Length vs Petal Width')
plt.show()
```

```python
#We can clearly see the grouping in the plots with the red dots, which correspond to species Setosa. 
#The green and blue dots are not so clearly separable.
#From this we can also take the number of clusters as '3'
```

## Useing the KMeans algorithm to see if it can create the clusters automatically.

```python
model = KMeans(n_clusters=3)
model.fit(x)
```

```python
#model.labels_ holds the array of the cluster ids
print (model.labels_)
```

```python
#To verify the output of the model, letâ€™s visualise it

colors = np.array(['red', 'green', 'blue'])

# The fudge to reorder the cluster ids.
predictedY = np.choose(model.labels_, [1, 0, 2]).astype(np.int64)

# Plot the classifications that we saw earlier between Petal Length and Petal Width

plt.scatter(x['Petal Length'], x['Petal Width'], c=colors[y.Target], s=40)
plt.title('Before classification')
plt.show() 
```

```python
# Plot the classifications according to the model

plt.scatter(x['Petal Length'], x['Petal Width'], c=colors[y.Target], s=40)
plt.title("Model's classification")
plt.show()
```
